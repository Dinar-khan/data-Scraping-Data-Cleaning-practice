### put everything together

```python
brand_ny =[]
model_ny =[]
mileage_ny =[]
price_ny =[]
year_ny =[]


for result in results:
    # brand
    try:
        brand_ny.append(result.find('span',{'class':'truncate'}).text)
    except:
        brand_ny.append('')
    # model
    try:
        model_ny.append(result.find('div',{'class':'truncate text-xs'}).text)
    except:
        model_ny.append('')
    # mileage
    try:
        mileage_ny.append(result.find('div',{'class':'flex w-full justify-between'}).get_text().strip(''))
    except:
        mileage_ny.append('')
    # year
    try:
        year_ny.append(result.find('span',{'class':'vehicle-card-year text-xs'}).get_text())
    except:
        year_ny.append('')
    # price
    try:
        price_ny.append(result.find('div',{'class':'heading-3 my-1 font-bold'}).text)
    except:
        price_ny.append('')
```

### Data frame creation --one page

```python
data_single_ny = pd.DataFrame({'Brand':brand_ny,'Model':model_ny,'Mileage':mileage_ny, 'Year':year_ny, 'Price':price_ny})
```
